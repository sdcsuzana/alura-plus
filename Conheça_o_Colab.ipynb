```
# Create a dictionary with English to Portuguese column name translations
column_translations = {
    'work_year': 'ano',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_empregado',
    'remote_ratio': 'taxa_remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'tamanho_empresa'
}

# Rename the columns using the dictionary
df.rename(columns=column_translations, inplace=True)

# Display the updated column names to confirm the translation
print(df.columns)
```
