```
# Create a dictionary to translate the categories in 'remoto'
remote_translations = {
    0: 'Presencial',
    50: 'Hbrido',
    100: 'Remoto'
}

# Replace the values in the 'remoto' column with their Portuguese translations
df['remoto'].replace(remote_translations, inplace=True)

# Display the updated value counts for the 'remoto' column
display(df['remoto'].value_counts())
```
